```python
from faker import Faker
import json
import requests

# Initialize the Faker instance
faker = Faker()

# Initialize a counter for sequential usernames and emails
counter = 1

def generate_customer_data():
    """
    Generate a single customer data dictionary.
    """
    global counter
    username = f"kanchCust{counter}"
    email = f"kanchcust{counter}@gmail.com"
    
    # Increment the counter for the next customer
    counter += 1
    
    return {
        "username": username,
        "first_name": faker.first_name(),
        "last_name": faker.last_name(),
        "address": faker.street_address(),
        "city": faker.city(),
        "state": faker.state(),
        "zip_code": faker.zipcode(),
        "phone_number": "".join([str(faker.random_digit()) for _ in range(10)]),  # Generate 10 digits
        "email": email,
        "credit_card": faker.credit_card_number(card_type="mastercard"),
        "password": "TestPass@123"  # Common password
    }

# Generate 10,000 customer records
customers = [generate_customer_data() for _ in range(10000)]

# Divide records into batches of 500
batch_size = 500
batches = [customers[i:i + batch_size] for i in range(0, len(customers), batch_size)]

# API endpoint URL
api_url = "http://127.0.0.1:8000/api/customers/bulk-register/"

# Headers for the API request
headers = {
    "Content-Type": "application/json",
    # Replace 'csrftoken' with the correct CSRF token or Authorization header as required
}

# Function to send a batch of customers to the API
def send_batch_to_api(batch):
    """
    Send a batch of customers to the API.
    """
    try:
        response = requests.post(api_url, headers=headers, data=json.dumps(batch))
        if response.status_code == 201:
            print(f"Batch of {len(batch)} records processed successfully.")
        else:
            print(f"Failed to process batch. Status code: {response.status_code}")
            print(f"Response: {response.text}")
    except Exception as e:
        print(f"An error occurred while sending the batch: {str(e)}")

# Iterate over the batches and send them to the API
for i, batch in enumerate(batches):
    print(f"Processing batch {i+1}/{len(batches)}...")
    send_batch_to_api(batch)

print("All batches have been processed.")
```
